# Customer Success Playbooks - Services & Contract Risk Segments

## 📊 PLAYBOOK OVERVIEW

These playbooks translate service and contract churn insights into specific CS team actions. Each playbook includes trigger conditions, risk levels, intervention strategies, and success metrics.

A critical note before deploying these playbooks: the predictive model that underpins account prioritisation achieves 81.9% accuracy, 68.7% precision, and 58.2% recall. This means the model misses approximately 4 in every 10 real churners. CSMs should treat model outputs as a starting point, not a final verdict. Personal knowledge of each account — support ticket history, sentiment from past calls, engagement patterns, renewal history — must complement the model score for accurate prioritisation. The CSM who knows their accounts will always outperform the model alone.

Additionally, two prioritisation tools exist for these segments and should be used together:

- **Health Score Priority List:** Ranks accounts by churn risk. Useful for identifying which customers are most statistically likely to churn.
- **Revenue-Weighted Priority List:** Ranks accounts by annual revenue at risk (churn probability × monthly charges × 12). Useful for identifying which accounts to call first when CSM capacity is limited.

A customer ranked 15th on the health score list may rank 1st on the revenue list if they pay $105/month. Always cross-reference both lists before planning your weekly outreach.

## 🚨 PLAYBOOK 1: Fiber Optic Premium Retention Program
**PRIORITY: CRITICAL (Highest Revenue Impact)**

### Risk Profile
- **Trigger Condition:** InternetService = Fiber Optic
- **Risk Level:** 41.9% churn rate (highest of all service types)
- **Affected Population:** 3,096 customers (44% of base)
- **Average Monthly Charge:** $91.50 (highest across all segments)
- **Annual Revenue at Risk:** $1,370,000+

### Why This Segment is Different

Fiber Optic customers are the company's highest-paying customers. At `$91.50` average monthly charges — nearly double what DSL customers pay at `$58` — every Fiber Optic customer lost carries a disproportionately large revenue impact. This is not just a churn problem. It is a perceived value problem.

Customers paying a premium have higher expectations. When those expectations are not met — whether through service quality, support responsiveness, or the feeling that the product is not worth what they pay — churn follows almost inevitably. Standard CS treatment is not sufficient for this segment. Every touchpoint needs to reflect the premium nature of the relationship.

### CS Ownership
- **Primary:** Senior CSM (minimum 2 years experience)
- **Secondary:** Technical Account Support Specialist
- **Escalation:** CS Manager for accounts showing 2+ warning signals simultaneously

### Intervention Timeline

**Onboarding Phase (Days 1-30)**
```
Week 1:
□ Premium welcome call within 24 hours (Senior CSM, not junior)
□ Service quality baseline assessment (speeds, reliability, setup)
□ Success plan creation aligned to customer's specific use case
□ Technical setup review — confirm installation meets expectations
□ Set monthly check-in cadence from Day 1

Week 2:
□ First service quality check-in
□ Confirm customer is experiencing advertised speeds
□ Identify any early frustrations or unmet expectations
□ Document customer's primary use case (streaming, work from home, gaming)

Week 3:
□ Usage and satisfaction check-in
□ Flag any support tickets raised in first 3 weeks
□ Introduce premium features relevant to their use case
□ Proactive: "Is the service performing the way you expected?"

Week 4:
□ 30-day premium service review
□ NPS survey sent
□ Confirm perceived value: "Is Fiber Optic meeting your needs?"
□ Document any gaps between expectation and experience
```

**Adoption and Retention Phase (Days 31-180)**
```
Monthly touchpoints:
□ Month 2: Service performance review + feature adoption check
□ Month 3: First Quarterly Business Review (QBR) — focus on value delivered
□ Month 4: Renewal conversation planted
□ Month 5: Expansion discussion (add-ons, upgrades)
□ Month 6: Mid-year satisfaction review
```

**Ongoing Retention Phase (Days 181+)**
```
Quarterly touchpoints:
□ QBRs with documented value realisation
□ Service performance benchmarking (compare to promised speeds)
□ Proactive contract renewal planning (90 days before renewal)
□ Expansion and upgrade opportunities
```

### Early Warning System Triggers
Immediate Senior CSM outreach required if:
- ⚠️ Service complaint or technical ticket raised (respond within 4 hours)
- ⚠️ 2+ support tickets in 30 days
- ⚠️ Speed or reliability complaint (service quality is their primary pain point)
- ⚠️ No engagement with CSM outreach for 14+ days
- ⚠️ Downgrade inquiry or contract cancellation request
- ⚠️ Health score drops below 40 (Critical zone)
- ⚠️ Revenue at risk exceeds $800 annually (appears in top 20 of revenue-weighted list)

### Perceived Value Intervention Framework

The central challenge with Fiber Optic churn is perceived value. When a customer feels the service does not justify $91.50/month, no amount of relationship management fully compensates. CS must address both the relationship AND the value gap.

**Value Gap Signals:**
- Customer mentions competitor pricing
- Customer asks "what am I actually getting for this price?"
- Customer compares their experience to a cheaper alternative
- Support tickets about service not meeting expectations

**Value Gap Responses (tiered by severity):**

Tier 1: Value Reaffirmation (CSM)
- Conduct value realisation review — document what the customer has gained
- Walk through premium features they may not be using
- Show performance data: actual speeds, uptime, reliability metrics
- Connect service quality to their specific use case (working from home, streaming quality)

Tier 2: Service Quality Review (CSM + Technical)
- Escalate to technical team for service audit
- Confirm installation quality and router placement
- Offer free technical health check visit if available
- Provide temporary service credit for documented service failures

Tier 3: Retention Offer (Manager approval required)
- 3-month loyalty discount (up to 15%)
- Premium add-on features at no extra charge for 6 months
- Priority technical support queue
- Account review with CS Manager to demonstrate company commitment

### Success Metrics
**Leading Indicators:**
- Service satisfaction score: >8.5/10 (higher bar than general population — premium customers expect more)
- Support ticket resolution time: <4 hours (same day)
- Feature adoption: 3+ features beyond base Fiber Optic
- Response rate to CSM outreach: >85%

**Lagging Indicators:**
- 12-month Fiber Optic retention: >75% (vs current 58%)
- NPS from Fiber Optic segment: >7
- Average Fiber Optic tenure: increase by 6 months over 12 months
- Expansion revenue from Fiber Optic segment: track quarterly

### Financial Impact
**Current State:**
- Fiber Optic customers churning: ~1,297 annually
- Annual revenue lost: $1,370,000+

**Target State:**
- Reduce Fiber Optic churn from 41.9% to 30%
- Customers saved: ~370 per year
- Revenue retained: $407,929 annually
- Program cost estimate: $90,000 (senior CSM capacity + technical support)
- **Net benefit: $317,929**
- **ROI: 253%**

## 📋 PLAYBOOK 2: Month-to-Month Contract Conversion Program
**PRIORITY: CRITICAL (Highest Volume Problem)**

### Risk Profile
- **Trigger Condition:** Contract = Month-to-month
- **Risk Level:** 42.7% churn rate (15x higher than two-year contracts at 2.8%)
- **Affected Population:** 3,875 customers (55% of base)
- **Average Monthly Charge:** $66.40
- **Annual Revenue at Risk:** $1,318,674

### Why This Segment is Different

Month-to-month customers represent the company's lowest commitment segment. The data is unambiguous: a customer on a two-year contract churns at 2.8%. The same customer on a month-to-month contract churns at 42.7%. The contract type is not just a billing preference — it is a proxy for how invested the customer feels in the relationship.

The CS strategy for this segment is dual-track. First, reduce churn by demonstrating value and building loyalty. Second, actively convert month-to-month customers to longer-term contracts. Every customer successfully converted is effectively removed from the highest-risk pool permanently.

### CS Ownership
- **Primary:** Standard CSM assignment
- **Contract Conversion Specialist:** Dedicated team member focused on upgrade conversations
- **Escalation:** CS Manager for accounts declining conversion after two attempts

### Intervention Timeline

**Early Engagement Phase (Days 1-60)**
```
Week 1:
□ Standard onboarding call
□ Understand why customer chose month-to-month (flexibility? price testing? distrust?)
□ Document reason — this determines the conversion approach
□ Set 30-day check-in

Week 2-4:
□ Monitor usage patterns
□ Flag any early friction or support issues
□ Ensure customer is experiencing core product value

Days 31-60:
□ First value realisation check-in
□ Plant the contract conversation: "Have you thought about locking in your rate long-term?"
□ Share contract benefits without pressure
□ Document customer response for follow-up
```

**Conversion Phase (Days 61-180)**
```
Month 3:
□ First structured contract conversion conversation
□ Present financial case: "Annual plan saves you $X per year"
□ Address flexibility concern: "What would need to be true for you to commit long-term?"
□ Offer trial period reassurance: "You've now seen the product works for you"

Month 4-5:
□ Follow up on conversion conversation
□ Introduce time-sensitive incentive if available
□ Escalate to manager if conversion is stalling

Month 6:
□ Final conversion push before customer is considered entrenched in month-to-month pattern
□ If not converting: shift strategy to loyalty building and churn risk monitoring
```

**Ongoing Monitoring Phase (Days 181+)**
```
Monthly:
□ Health score review (month-to-month customers should be checked monthly, not quarterly)
□ Revenue-weighted priority check: are they appearing in top 20 at-risk accounts?
□ Renewal reminder sent 30 days before each monthly renewal
□ Conversion opportunity revisited at 6-month intervals
```

### Contract Conversion Playbook

The conversion conversation is the most important CS action for this segment. It requires preparation and the right framing.

**Before the Conversion Call:**
- Review customer's usage data: are they getting value?
- Check support ticket history: any unresolved frustrations?
- Calculate the customer's specific savings: personalise the financial case
- Review their health score and revenue at risk figure
- Check if they appear on the revenue-weighted priority list

**The Conversion Conversation Framework:**
```
Step 1: Value Confirmation (2 minutes)
"Before we talk about anything else — how has the service been for you?"
"What's been working well?"
[Document positives — these are your conversion ammunition]

Step 2: Future-Pacing (2 minutes)
"It sounds like [specific value point] has been really useful for you."
"What do you see yourself using the service for over the next year or two?"
[Get the customer talking about their future with the product]

Step 3: The Transition (1 minute)
"Given that you're planning to keep using the service, it might be worth looking at whether a longer-term plan makes sense for you financially."
[Soft, not pushy — framing it as their financial interest, not a sales push]

Step 4: The Financial Case (3 minutes)
"Customers on our annual plan typically save $X compared to month-to-month."
"For you specifically, based on what you're paying now, that works out to about $Y per year."
[Use real numbers — personalised savings land better than generic percentages]

Step 5: Address the Flexibility Concern (if raised)
"I understand wanting to keep your options open — that's exactly why a lot of customers start month-to-month."
"Most customers who switch tell me they wished they'd done it sooner once they saw the savings."
[Normalise the decision without dismissing the concern]

Step 6: Close or Set Follow-Up
"Would you like me to put together the details for an annual plan for you to review?"
OR
"Let's revisit this in 30 days — I'll send you the numbers in the meantime."
```

### Retention Offers (for at-risk month-to-month customers)
**Tier 1: Conversion Incentive (CSM approved)**
- First month free on annual plan
- Rate lock guarantee (protection from future price increases)
- Priority support queue for first 3 months after conversion

**Tier 2: Loyalty Incentive (Manager approved)**
- 10% discount for 12-month commitment
- Free add-on service for first 6 months
- Dedicated CSM for high-value accounts converting to annual

**Tier 3: Retention Offer for Non-Converters (Manager approved)**
- Churn risk discount: 1-month credit to prevent cancellation
- Service upgrade at current price
- Escalation call with CS Manager

### Success Metrics
**Leading Indicators:**
- Month-to-month to annual conversion rate: target 20% within 6 months
- Response rate to conversion conversation: >75%
- Month-to-month customer NPS: >6 (lower bar — these customers are more transactional)

**Lagging Indicators:**
- Month-to-month churn reduction: from 42.7% to 30% (target)
- Annual contract upgrade rate: 15% of month-to-month base per quarter
- Revenue recovered through conversion: track monthly

### Financial Impact
**Current State:**
- Month-to-month customers churning: 1,655 annually
- Annual revenue lost: $1,318,674

**Target State:**
- Reduce month-to-month churn from 42.7% to 30%
- Customers saved through direct retention: 492
- Additional customers removed from risk pool through conversion: estimated 250
- Revenue retained: $392,000+ annually
- Program cost estimate: $65,000
- **Net benefit: $327,000**
- **ROI: 503%**

## 💳 PLAYBOOK 3: Electronic Check Friction Resolution Program
**PRIORITY: HIGH (Highest Efficiency Intervention)**

### Risk Profile
- **Trigger Condition:** PaymentMethod = Electronic check
- **Risk Level:** 45.3% churn rate (highest of all payment methods)
- **Affected Population:** 2,365 customers (33.6% of base)
- **Average Monthly Charge:** $76.26 (second highest of all payment methods)
- **Annual Revenue at Risk:** $980,040

### Why This Segment is the Highest Efficiency Intervention

This segment carries a unique strategic advantage that no other segment has: 78.2% of electronic check customers are also on month-to-month contracts. This means a single intervention addressing electronic check customers simultaneously addresses the month-to-month churn problem for the majority of this group.

Addressing both Playbook 2 and Playbook 3 together for this overlapping population is the most efficient use of CS resources in the entire portfolio. One outreach call can progress both the payment friction resolution and the contract conversion conversation simultaneously.

Additionally, electronic check customers pay an average of $76.26/month — higher than mailed check ($43.92) and close to automatic payment methods. This means many of these customers are on higher-tier services, likely Fiber Optic, creating a three-way overlap: electronic check, month-to-month, and Fiber Optic. Customers sitting at all three intersections should be the absolute first priority on any outreach list.

### CS Ownership
- **Primary:** Standard CSM
- **Product Collaboration:** CS to escalate payment friction patterns to Product and Development teams
- **Operations:** Billing support team for payment method migration assistance

### The Core Hypothesis

Electronic check churn is driven by two compounding factors. First, payment friction — manual electronic processes are more prone to failed transactions, confusing interfaces, and technical issues that erode trust. Second, low commitment — the same customer profile that chooses electronic check over automatic payment also tends to choose month-to-month over annual contracts. These are customers who have not fully committed to the relationship.

The intervention therefore works on two levels: reduce payment friction through better onboarding and product improvements, and increase commitment through contract conversion.

### Intervention Timeline

**Onboarding Phase (Days 1-30)**
```
Week 1:
□ Standard welcome call
□ Payment method setup verification: "Has your first payment processed successfully?"
□ Walkthrough of electronic check process (step-by-step, avoid assuming familiarity)
□ Provide written guide to payment process via email
□ Set up payment confirmation notifications

Week 2:
□ Confirm payment received and processed
□ Ask directly: "Have you had any issues with the payment process?"
□ Document any friction points

Week 3-4:
□ Introduce automatic payment option as a convenience (not a pressure)
□ "A lot of customers switch to automatic payment to avoid having to remember each month — would that be useful for you?"
□ If interest shown: initiate payment migration conversation
```

**Friction Resolution Phase (Days 31-90)**
```
Month 2:
□ Payment confirmation check-in after second payment cycle
□ Document any failed payment attempts or processing issues
□ If issues identified: escalate to billing support immediately
□ Re-introduce automatic payment migration if first attempt declined

Month 3:
□ Three-month payment history review
□ If clean payment history: use as trust signal for contract conversion
□ If payment issues: escalate to product team — this is a systemic friction point
```

**Ongoing Monitoring Phase (Days 91+)**
```
Monthly:
□ Flag any failed payment events in customer record
□ Monitor for downgrade or cancellation requests following payment failures
□ Quarterly: revisit automatic payment migration for customers who declined earlier
□ Track: what percentage of electronic check customers successfully migrate to automatic?
```

### Payment Migration Conversation Framework

The goal is to move customers from electronic check to automatic payment (bank transfer or credit card). Automatic payment customers churn at 15-17% — less than a third of the electronic check rate.

```
Frame it as a convenience, not a company benefit:
"I wanted to mention that a lot of customers on electronic check end up switching to automatic payment 
just to make their lives easier — it means you never have to remember to process a payment 
and you avoid any risk of a failed transaction affecting your service."

If they ask why automatic payment is better:
"The main thing is reliability. Automatic payments go through consistently and you get a 
confirmation each month. Electronic check can sometimes have processing delays depending on 
your bank, and we don't want that to ever cause an interruption to your service."

If they raise concerns about security:
"That's a fair question. Bank transfer is actually one of the most secure methods — 
it processes directly from your account without storing your card details. 
We can walk through how it works if that would help."
```

### Product and Development Escalation

CS should not only solve electronic check churn at the account level. The payment friction issue is systemic and requires a product-level response. CS leadership should compile the following from CSM interactions and escalate to the Product team quarterly:

```
Escalation Report: Electronic Check Friction
□ Volume of payment failures in the quarter
□ Most common failure reasons (insufficient funds, processing errors, bank rejections)
□ Customer feedback on payment interface complexity
□ Comparison: support ticket volume for electronic check vs automatic payment customers
□ Recommendation: automation roadmap for electronic check processing
```

The long-term solution is to make electronic check as frictionless as automatic payment. Until that product improvement is delivered, CS manages the gap through proactive outreach.

### Retention Offers
**Tier 1: Payment Migration Incentive (CSM approved)**
- First month free when switching to automatic payment
- Payment confirmation setup assistance (white-glove for high-value accounts)

**Tier 2: Retention Offer for At-Risk Electronic Check Customers (Manager approved)**
- 1-month service credit following documented payment failure
- Priority billing support access
- Temporary payment plan flexibility for accounts in financial difficulty

**Tier 3: Combined Electronic Check + Month-to-Month Retention (Manager approved)**
- Bundle: switch to automatic payment AND annual contract → 2 months free
- Rate lock + automatic payment setup assistance
- Dedicated CSM for first 6 months of new contract

### Success Metrics
**Leading Indicators:**
- Electronic check to automatic payment migration rate: target 30% within 6 months
- Failed payment rate: track monthly (benchmark before and after intervention)
- Response rate to payment migration conversation: >70%

**Lagging Indicators:**
- Electronic check churn reduction: from 45.3% to 33% (target)
- Automatic payment adoption in this segment: track quarterly
- Product escalation: number of friction issues logged and resolved

### Financial Impact
**Current State:**
- Electronic check customers churning: 1,071 annually
- Annual revenue lost: $980,040

**Target State:**
- Reduce electronic check churn from 45.3% to 33%
- Customers saved: ~236 per year
- Revenue retained: $216,426 annually
- Efficiency bonus: simultaneous progress on month-to-month conversion for 78.2% of this segment
- Program cost estimate: $45,000
- **Net benefit: $171,426**
- **ROI: 381%**

## 🔒 PLAYBOOK 4: Add-On Feature Adoption Program
**PRIORITY: HIGH (Prevention-Focused)**

### Risk Profile
- **Trigger Condition:** OnlineSecurity = No AND/OR TechSupport = No (for internet-connected customers)
- **Risk Level:** ~42% churn rate for both (nearly 3x higher than customers with add-ons at 14-15%)
- **Affected Population:** Significant overlap with Fiber Optic and month-to-month segments
- **Note:** Customers with "No internet service" are excluded — their 7.4% churn rate reflects a different, more stable segment and should not be conflated with internet users lacking add-ons

### Why This Segment is Different

Customers without Online Security and Tech Support are not necessarily aware they are at higher risk. Unlike contract type or payment method, they may not have made a conscious decision to go without these features. Many simply were not given a compelling reason to adopt them during onboarding.

This makes the intervention fundamentally different from the other three playbooks. Here, CS is not primarily in retention mode. CS is in education and adoption mode. If a customer adopts Online Security and Tech Support, their churn risk drops from 42% to approximately 15%. The most powerful retention action is proactive feature adoption, not reactive outreach after the customer is already considering leaving.

### CS Ownership
- **Primary:** Onboarding CSM (adoption must be addressed early)
- **Secondary:** Product Adoption Specialist for feature education
- **Collaboration:** Product team for feature trial programmes

### The Two Root Causes

**Root Cause 1: Customers feel vulnerable without Online Security**
Customers without Online Security are exposed to online threats, fraud, and scams. When these incidents occur, they blame the service provider — even if the provider is not technically responsible. The frustration and distrust that follows accelerates churn.

**Root Cause 2: Customers without Tech Support face unresolved frustration**
Complex product features, technical errors, and setup difficulties are inevitable. Without dedicated Tech Support, customers either struggle in silence or raise tickets through standard support channels where resolution is slower. Accumulated frustration from unresolved technical issues is a primary churn driver.

### Intervention Timeline

**Onboarding Phase — Feature Adoption Window (Days 1-30)**

The first 30 days are the highest-leverage adoption window. Customers are most open to guidance at this stage and have not yet formed habits that exclude these features.

```
Week 1:
□ During welcome call: assess current Online Security setup
□ "Do you have any antivirus or security software currently?"
□ If no: introduce Online Security feature with specific use case
□ "A lot of our customers on [internet type] use Online Security to protect against phishing and fraud"
□ Offer free trial if available

Week 2:
□ Technical setup check-in
□ "Have you had any questions about how to use any of the features?"
□ "If you ever run into a technical issue, are you aware that Tech Support is available?"
□ If not subscribed: demonstrate Tech Support value with a specific example

Week 3-4:
□ Follow up on any features introduced in Week 1-2
□ If Online Security trial active: check experience
□ If Tech Support not subscribed: share a relevant scenario
□ "Last week one of our customers had [common issue] — Tech Support resolved it in 20 minutes. Worth having that available?"
```

**Adoption Deepening Phase (Days 31-90)**
```
Month 2:
□ Feature adoption check: which features are actively used?
□ For unsubscribed features: one more value demonstration
□ If customer had any security concern or technical issue in Month 1: use as direct case for adoption

Month 3:
□ 90-day adoption review
□ Customers who adopted features: celebrate their usage and document value
□ Customers who did not adopt: acknowledge and note in account record — do not pressure
□ Flag non-adopters with active Fiber Optic service for additional attention
```

**Ongoing Monitoring Phase (Days 91+)**
```
Quarterly:
□ Review feature adoption status in account
□ Any security incidents or technical issues in the quarter? Use as adoption conversation trigger
□ New feature releases relevant to this customer? Introduce proactively
```

### Feature Adoption Conversation Framework

**For Online Security:**
```
Lead with protection, not product:
"One thing I always mention to customers on [internet service] is Online Security. 
With the amount of online activity most people have today — banking, shopping, email — 
having a layer of protection against phishing and fraud is something most customers 
tell us they wish they'd set up earlier."

If they're hesitant on cost:
"We do offer a trial period so you can experience it before committing. 
Would that be useful to try?"
```

**For Tech Support:**
```
Lead with peace of mind:
"Tech Support is one of those things customers don't think they need until they need it. 
If you ever run into a technical issue — router setup, connection problems, device questions — 
having dedicated Tech Support means you get a specialist rather than waiting in a general queue."

Specific framing for Fiber Optic customers:
"Given that you're on Fiber Optic, which has more technical components than a basic connection, 
having Tech Support on hand is something a lot of our premium customers rely on."
```

### Retention Offers
**Tier 1: Feature Trial (CSM approved — no cost)**
- 30-day Online Security trial (convert to paid if adopted)
- 30-day Tech Support trial during onboarding
- Feature education session: 1-hour walkthrough

**Tier 2: Adoption Bundle (Manager approved)**
- 3-month discount on Online Security + Tech Support bundle
- Bundle with annual contract for combined savings
- Free security audit for Fiber Optic customers

### Success Metrics
**Leading Indicators:**
- Feature adoption rate within 90 days: target 35% of non-adopters
- Trial-to-paid conversion rate: target 60% of trial users
- Feature education session attendance: track monthly

**Lagging Indicators:**
- Churn rate for non-adopters vs adopters: track quarterly (target: adoption closes the gap from 42% toward 20%)
- Support ticket volume comparison: adopters vs non-adopters
- 12-month retention comparison: customers who adopted Online Security or Tech Support vs those who did not

### Financial Impact

The financial case for this playbook is different from the others. Rather than recovering revenue from active churners, this playbook prevents future churn through early intervention. The return compounds over time as adopted customers stay longer.

**Conservative Estimate:**
- Target population: 2,000+ internet customers without key add-ons
- Target adoption rate: 35% (700 customers)
- Expected churn reduction for adopters: from 42% to 20% (22 percentage points)
- Customers saved through adoption: ~154 annually
- Average monthly charge: $65
- **Annual revenue protected: ~$120,120**
- Program cost: $30,000 (adoption specialist time + trial programme)
- **Net benefit: $90,120**
- **ROI: 300%**
