## 1. What is Customer Churn?

Every customer relationship has a moment of truth — the point where a customer decides whether the value they're receiving justifies staying. In telecommunications, when that answer becomes "no", the customer churns. They cancel, they switch, or they simply stop engaging.

Churn is fundamentally a relationship failure. It means a customer reached out for help and wasn't heard, encountered a problem that wasn't solved, or never fully understood the value of what they were paying for. The data shows the outcome — but behind every churned customer is a series of moments where the company had an opportunity to intervene and didn't.

In the highly competitive telecom market, where switching barriers are low and alternatives are plentiful, the Customer Success team is the primary defence against churn. This project exists to give that team the data, tools, and playbooks they need to identify at-risk customers before they reach that moment of truth — and intervene in time to change the outcome.

## 2. Why Churn Matters in Telecommunications
Telecommunications is one of the most difficult industries in which to build lasting customer loyalty. Switching a provider takes minutes. Competitors are aggressive. And most customers cannot tell the difference between one provider's service and another until something goes wrong.

This creates a specific challenge for CS teams: the window to demonstrate value and build loyalty is short, the triggers for leaving are often invisible until it is too late, and the financial consequences of losing a customer compound quickly.

### What a Churned Customer Actually Costs
The immediate cost is the lost monthly revenue. But that is only part of the picture. A customer who leaves after six months never reaches the point where they refer others, expand their plan, or advocate for the product. The company then has to spend five to ten times the cost of retention just to replace them through new acquisition.

For a CS team, this means every saved customer is worth significantly more than their monthly charge suggests. A Fiber Optic customer paying `$91.50` per month who stays for three years instead of six months generates over `$3,000` in revenue. The same customer churning at month six generates `$549`. That gap — `$2,500` or more — is what proactive CS work is actually protecting.

### The Segment That Matters Most
Not all customers carry equal risk or equal value. The table below shows how contract type maps to both lifetime value and churn risk in this dataset:

| Customer Segment | Typical LTV | Churn Risk |
|---|---|---|
| **Premium (2-year contracts)** | `$3,000 to $8,000+` | Low (8 to 12%) |
| **Standard (1-year contracts)** | `$1,200 to $3,000` | Medium (15 to 20%) |
| **Basic (month-to-month)** | `$300 to $800` | High (35 to 50%) |
| **Low-engagement** | Under `$200` | Very High (over 50%) |

The implication for CS is practical: time spent retaining a two-year contract customer has a fundamentally different return than time spent on a month-to-month customer at risk of leaving after 30 days. Prioritisation is not optional — it is the core skill.

### The 80/20 Reality
In most telecom businesses, 20% of customers generate 80% of revenue. Within that 20%, churn risk is lower but the stakes per account are higher. A single churned premium customer can erase the revenue gain of retaining five lower-value accounts. CS teams that lack data tools tend to spread their effort evenly. CS teams with health scores and priority lists focus where the revenue impact is highest.

This project was built to give the CS team exactly that capability.

## 3. Business Problem Statement

This telecom company is experiencing a **26.5% annual churn rate**, representing **1,869 customers** who have left the service. At an estimated average Customer Lifetime Value of `$1,500-$2,000`, this represents **`$2.8M - $3.7M` in lost revenue potential**. However, the analysis of actual monthly charges across all 1,869 churned customers reveals a precise annual revenue loss of `$1,452,475`. The LTV-based estimate of `$2.8M to $3.7M` reflects the long-term relationship value lost, not the immediate annual billing impact."

Currently, the company lacks:
- **Predictive capability** to identify at-risk customers before they churn
- **Visibility** into the key drivers causing customers to leave
- **Strategic framework** for prioritizing retention efforts
- **Data-driven insights** to guide product and service improvements

### Analysis Objectives

This analysis aims to:
1. **Identify** the top 3-5 factors driving customer churn
2. **Build** a predictive model to score individual customer churn risk
3. **Segment** customers by churn probability and value
4. **Quantify** the revenue impact and ROI of retention interventions
5. **Recommend** data-driven retention strategies for the Customer Success team
6. **Create** an actionable intervention framework for proactive customer management

### Stakeholder Impact

**Customer Success Team:**
- Prioritize outreach to high-risk, high-value customers
- Deploy targeted retention campaigns
- Measure intervention effectiveness

**Product Team:**
- Address service gaps and pain points
- Improve onboarding and engagement
- Enhance feature adoption

**Executive Leadership:**
- Quantify financial impact of churn
- Make data-driven investment decisions
- Track retention KPIs against targets

## 4. Success Metrics
### Business Success Criteria
The primary measure of success for this project is not model accuracy — it is whether the CS team can use the outputs to retain more customers and recover lost revenue.

**Short-term (3 to 6 months):**
- Reduce churn rate by 3 to 5 percentage points, from 26.54% toward 21 to 23%
- Retain an additional `$400K to $600K` in annual revenue through targeted CS interventions
- Achieve a 15 to 20% save rate on proactive outreach campaigns

**Medium-term (6 to 12 months):**
- Reduce churn to below the industry average of 22%
- Improve Net Promoter Score (NPS) by 10 or more points
- Increase Customer Satisfaction (CSAT) scores by 8 to 12%

**Long-term (12 to 24 months):**
- Establish proactive retention as a core CS operational capability
- Achieve an LTV to CAC ratio of 4:1 or higher
- Build a scalable early warning system that CS teams can run without analyst support

### Analytical Success Criteria
To deliver those business outcomes, the predictive model needs to perform well enough to be useful in practice. The model built in this project achieved the following:

- **Accuracy: 81.9%** — the model correctly predicts churn or retention for 4 in every 5 customers
- **Precision: 68.7%** — when the model flags a customer as at-risk, it is correct approximately 7 times out of 10, reducing wasted CSM effort on false alarms
- **Recall: 58.2%** — the model correctly identifies 58% of customers who will churn, meaning CSMs should complement model outputs with their own customer knowledge to catch the remainder

Precision fell short of the initial 75% target. However, at 68.7%, the model still provides a significantly more reliable at-risk list than manual identification alone, and the human-in-the-loop approach compensates for the gap in recall.

## 5. Connection to Real-World Experience

During my tenure as Associate Product Manager at **Peadato Limited**, I directly experienced the impact of customer churn on business growth. Our platform initially focused on sales-driven customer acquisition, but we observed that:

- **Early customer frustration** led to rapid churn within the first 30-90 days
- **Lack of product understanding** created unrealistic expectations and disappointment
- **Poor onboarding experience** resulted in low engagement and eventual abandonment

**The Pivot:**
By shifting from a sales-first approach to a **user-centered education model**, we:
- Reduced early-stage churn by approximately **30%**
- Improved customer satisfaction from **65% to 95%**
- Increased product adoption among SMEs by **60%**

**Key Lesson:**
Understanding the **"why" behind churn** is as important as the "what." Customers don't leave because of a single reason—they leave because of accumulated friction points, unmet expectations, and better alternatives. Data can reveal these patterns, but empathy and customer-centric design solve them.

This telecom churn analysis will apply those same principles: **combining quantitative pattern recognition with qualitative customer understanding** to create actionable, empathetic retention strategies.

## 6. Analytical Approach & Methodology
The analysis follows five phases. Each phase builds on the previous one. The end goal is not a report. It is a set of tools the CS team can use every day.

### Phase 1: Data Understanding and Preparation
Before any analysis, the data needs to be trustworthy. This phase covers loading and exploring 7,043 customer records, identifying data quality issues, and fixing them. Decisions made here affect the reliability of every finding that follows. A clean dataset is the foundation everything else depends on.

### Phase 2: Understanding Who Churns and Why
This phase answers the most important CS question: which customers are leaving and what do they have in common? The analysis looks at demographics, service types, contract structures, payment methods, and add-on adoption. Every finding is translated into a CS action. Not just "fiber optic customers churn more" but what the CS team should do about it and why it matters in revenue terms.

### Phase 3: Quantifying the Revenue Impact
Churn rates show the scale of the problem. Revenue figures show the urgency. This phase calculates the annual revenue loss associated with each churn driver. It also builds the business case for CS investment by showing which segments to prioritise first based on both churn risk and revenue value.

### Phase 4: Building the Prediction and Health Scoring Tools
This phase builds two tools for the CS team. First, a logistic regression model that scores every customer by churn probability. Second, a customer health scoring system that combines churn probability, tenure, and contract type into a single score from 0 to 100 with traffic light categories. Both tools are designed for daily CS use, not one-time analysis.

### Phase 5: Making the Findings Accessible
Analysis that stays in a notebook helps no one. This phase turns findings into formats the CS team can actually use. That includes CS playbooks for each at-risk segment, an executive business case with projected revenue recovery figures, and a four-page Power BI dashboard. The dashboard is designed so a CSM can open it every Monday morning and know immediately who to call first.
